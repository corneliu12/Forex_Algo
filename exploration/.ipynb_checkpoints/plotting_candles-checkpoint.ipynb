```python
import pandas as pd
import plotly.graph_objects as go
import datetime as dt
```

```python
pair = "EUR_USD"
granularity = "H4"
df = pd.read_pickle(f"../data/{pair}_{granularity}.pkl")
```

```python
df.tail()
```

```python
# getting first 100 candles to plot
df_plot = df.iloc[:100]
```

```python
# using plotly to graph candlestick charts
fig = go.Figure()
fig.add_trace(go.Candlestick(
    x=df_plot.time,
    open=df_plot.mid_o,
    high=df_plot.mid_h,
    low=df_plot.mid_l,
    close=df_plot.mid_c,
    line=dict(width=1), opacity=1,
    increasing_fillcolor='#24a06b',
    decreasing_fillcolor='#cc2e35',
    increasing_line_color='#2ec886',
    decreasing_line_color='#ff3a4c'
    
))
# changing colors of the grid lines and removing bottom slider
fig.update_xaxes(
    gridcolor="#1f292f",
    rangeslider=dict(visible=False)
)

fig.update_yaxes(
    gridcolor="#1f292f"    
)
# updating the layout of the plot
fig.update_layout(
    width=1680,
    height=600,
    margin=dict(l=10, r=10, b=10, t=10),
    paper_bgcolor="#2c303c",
    plot_bgcolor="#2c303c",
    font=dict(size=12, color="#e1e1e1")
)
    

fig.show()
```
